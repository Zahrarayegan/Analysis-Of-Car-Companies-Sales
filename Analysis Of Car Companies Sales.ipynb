```python
avg_prices = df_car.groupby('Manufacturer')['SalePrice'].mean().sort_values(ascending=False)
print(avg_prices)
```

```python
brand_counts = df_car['Manufacturer'].value_counts()
plt.Figure(figsize=(10,5))
sns.barplot(x=brand_counts.index, y=brand_counts.values, palette="coolwarm")
plt.title("Sales Count by Brand")
plt.xlabel("Car Brand")
plt.ylabel("Sales Volume")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.Figure(figsize=(10,10))
sns.barplot(x=avg_prices.index, y=avg_prices.values,palette="viridis")
plt.title("Average selling price by brand")
plt.xlabel("Manufacturer")
plt.ylabel("Average selling price")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_car['SaleDate'] = pd.to_datetime(df_car['SaleDate'])

df_car['saleYear'] = df_car['SaleDate'].dt.year

sales_by_year = df_car['saleYear'].value_counts().sort_index()

plt.Figure(figsize=(8,5))
sns.lineplot(x=sales_by_year.index, y=sales_by_year.values,markers='o', color='green')
plt.title('Sales per year')
plt.xlabel('year')
plt.ylabel('Sales volume')
plt.grid(True)
plt.show()
```
